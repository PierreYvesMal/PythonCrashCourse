# Variables
*Variables* are used to *store* information. The stored information can be retrieved using the name of the variable. In Python, the type is implied, so no need to declare a variable. Just use it directly.  
Variables can directly store the output of a mathematical operation and more.  
- You *write* a value to the variable using the *=* operator. It's always name of the variable on the left, new value to store on the right. You can store a new value on the variable at any time, this will overwrite the previous value.
- You *read* the stored value in a variable using the name of the variable directly.

```python
a = 3
print(a)

b = 1+1
print(b)

print(a+b)

c = a+b

print("The end result is:", c)
print("The end result is: " + str(c))
print(f"The end result is: {c}")
```

All three lines print the same thing.
- The first one is the fastest
- The second one is using "+", which is attempting to perform an operation betwen the left and right operand. However, these are not of the same type: a string (text) and an int (number). Since we are trying to *concatenate* text to print it, we first need to convert the number into text using the str() function.
- The third one is great because it directly creates a string in place. Meaning it's not up to the function to decide what to do with the different argument. Let's put that aside for now.

The print format can also be managed with string formatting

```python
hour = 11
minute = 8
print(f"The time is {hour:02d}:{minute:02d}")
pi = 3.141592653589793
print(f"Pi is approximately {pi:.2f}")
```

# Conditions
This is THE core concept of programming. It execute a code block under certain conditions.  

## If
Code block is executed only if the condition is *True*.  
if condition:  
  code to execute

```python
if True:
    print("This code block is executed because the condition is True")
```

We used *True* as the condition. The *if* keyword always expects a *boolean* type (True/False). Good thing: boolean is also the return type of a logical expression.  
Operators are used between left and right operand to create a logical expression.  
| Operator      | Description | Returns True if |
| ----------- | ----------- | --------- |
| ==      | equal       | Operands are the same |
| !=      | not equal       | Operands are the same |
| >   | Greater than        | Left is greater than right |
| >=   | Greater or equal to        | Left is greater or equal to |

```python
if 1 != 3:
    print("1 is different than 3")
```

The result of a logical operation can also be saved in a variable. The variable will be a boolean, that can be used for a condition.

```python
myVariable = 1==2
if myVariable:
    print(myVariable)
# Nothing is printed because 1 is not equal to 2.
```

## Else
*Else* block can optionnally follow an *if* block. The *else* block is executed when the *if* block is not.

```python
if myVariable:
    print("1 is equal to 2")
else:
    print("1 is not equal to 2")
print("The stored result of the logical 1==2 operation is ", myVariable)
```

## elif
Optionally follows an *if* or an *elif* block. Execute when:
- The if block was not executed
- The elif condition is met

An *else* block can also follow an *elif* block to represent "all other possibilies". It is a good practice to include the else block.

```python
myVar = 5-3
if myVar == 1:
    print("The result is 1")
elif myVar == 2:
    print("The result is 2")
elif myVar == 3:
    print("The result is 3")
else:
    print("The result is not 1,2, or 3. You'll have to admit this code is intellectually limited.")
```

Note that with conditions only you can technically already code basically anything (that doesn't interact with other hardware or software).  
However you'll die young if you just go like that. Imagine printing a count to 1000.

## Ternary expression
x if condition else y | this expressions is evaluated as X if the condition is true, as y otherwise.

```python
number = 13
print("even" if number % 2 == 0 else "odd")

print("even") if number % 2 == 0 else print("odd")

parity = "even" if number % 2 == 0 else "odd"
print(parity)
```

## Arguments
A function is often meant to behave differently based on what you give it. Just like the *print()* function prints something different, based on the text we give it.

Giving information to a function is called "Passing an argument". All arguments are present in the parenthesis, separated by the "," separator. When using the function, the number (and the type) of arguments must match the definition.

## Return
We use arguments to pass information to the function. From the function perspective, it is an "input". External world -> function

The opposite of that would be an "output". The function gives information to the rest of the code. It's called "returning". So the function returns something.

Let's create a function that computes the addition between two numbers, with and without return.

```python
#The function Add() computes the result of a+b.
def Add(a, b):
    result = a+b
```

This function is absolutely useless as it is, since the rest of the code cannot access the result, that only exists in the context of the function.

This is reffered as the *scope" of a variable. The variable only exists within its scope. So far, we had always created the variables in the main scope, so no such problem could occur.

If you create a variable within a scope (inside a block), it can only be accessed inside that block, and is destroyed while exiting the block. This works for functions but also for if/else, loops and more.

I'm not going to talk about global variables right now, because it should NOT used as a generic solving tool. Correct architecture is the way to go.

Note: this is worth whiteboard explanation if need be.

```python
Add(2,3)
print(result)
```

Instead, the function should have *returned* the result, so the caller has access to it.

```python
#The function Add() computes the result of a+b, AND returns the result to the caller
def Add(a, b):
    return a+b

result = Add(2,3) #Add() returns 5, so it is "replaced" by the value 5. In the end we have: result = 5

print(f"The result of 2+3 is {result}")
```

You can also specify the type of return. Again, this is useful as the return is seen when howevering the name of the function, wherever it is on the code.

```python
#The function Add() computes the result of a+b, AND returns the result to the caller
def Add(a : int, b : int) -> int:
    return a+b

result = Add(2,3) #Add() returns 5, so it is "replaced" by the value 5. In the end we have: result = 5

print(f"The result of 2+3 is {result}")
```

# Containers (or collections)
Containers meant to store multiple piece of related information at once. It's like the extension of a variable.

Let's see the most common ones

## list
This is the simplest one. It's a list of values. At any time, you add a value to the list, or retrieve one by its index.
- It is initialized with brackets [ ]
- Add/remove elements with append(), insert(), remove(), pop()
- You retrieve an element using its index [i]

```python
numbers = [1,5,-2]
numbers.append(0)
print(numbers)
print(numbers[1])
```

in most programming langages, including Python, containers are 0-indexed. Meaning we start counting positions from 0 and not one. So numbers[1] gave us the *second* element.

## Set
A set is like a list, without index, and where each element is unique
- Initialized with curley brackets {}
- Updated via add(), pop(), clear()
- CANNOT be retrieved using an index

```python
someSet = {1,5,-2, 5}
print(someSet) # See how the duplicate 5 is not included
```

## Tuple
Like a list, but of two objects only, *immutable*. You enforce a pair, like a coordinate.
- Initialized with parenthesis ()
- cannot be appened/removed (that's what meant by immutable)
- can be retrieved with [] or since it's only two elements, in a direct way (see below)

```python
point = (3,4)
x,y = point
print(f"({x},{y})")
print(point[0], ",", point[1])
```

## Dictionary
When you use a dictionnary, you want to find the definition using the word.

The word is the *key* and the definition is the *value* you are trying to access. You can create a form out of it where every field is a pair of key - value.  
It's a bit like a list of tuples.

- Initialized with curly brackets {} containing every *item* (key: value). Each item is separated by a coma ",", and follows key: value.
- Add and acces through brackets [] with the key

```python
user = {
    "name": "Alice",
    "age": 30
}

user["age"] = 31

print(user)

print(user["age"])

user["location"] = (32,40)

print(user["location"])
```

## Iterating on items of a container
Containers are often used in combination with loops to create them or to loop over their elements.

This is achieved by looping over
- the *indices* of the elements via the range() function
- the *elements* directly via the enumerate() function
- other particular iterable acquired via methods of the container itself (e.g. dict.keys())

```python
countdown = []
for i in range(0,5):
    countdown.append(i)

length = len(countdown) # Yeah you can use len() to get the number of elements of a container
for i in range(0,length):
    number = countdown[length-i-1]
    if number != 0:
        print(number)
    else:
        print("Bonne année !")

print("---")

for element in countdown: # This can't reverse the order
    print(element)
```

```python
# List
myList = [56, 43, -10]
for element in myList:
    print(element)

# Dictionnaries contain (key, value) pairs, so there are multiple ways to loop over them.
myDictionary = {"firstKey":"une valeur textuelle", "secondKey":4}
for key in myDictionary.keys():
    print(key)

for key, value in myDictionary.items():
    print(key, "->", value)
```

# Classes
You feel limited with your current types (int, string, bool, float, ...) ? Whatever your answer is, you need classes.

This lets you *define* a whole new type, and ways to interact with it.  
A class contains variables and functions that can only be accessed through an instance of that class (called an object).

Let's play God and define a Person.  
In the following, we define the person class, that contains
- The constructor *\_\_init__()*, a special function that defines *how* to create an object of type Person
- Two member variables "name" and "age". They do not make any sense of their own, they are accessed through a particular Person.
- Two functions that use the member variables.

```python
class Person:
    def __init__(self, name, age): #name and age are accessible only in the __init__ function. But member variables like self.name and self.age are available everywhere.
        self.name = name
        self.age = age

    def say_hello(self):
        print("Hello, my name is " + self.name)

    def get_older(self):
        self.age = self.age + 1
        print("I am now " + str(self.age) + " years old")
```

The code above only defines the class Person. We can now create a Person and interact with it.

```python
some_dude = Person("Jean-Jacques", 28) #This line creates a person. The __init__ function was automatically called with the given name and age.
some_dude.say_hello() #The say_hello MUST be accessed through a Person, via the dot "."
print(some_dude.age) #Member variables, defined inside the Person class must be accessed via a person aswell, but appart from that they are just normal variables.
some_dude.get_older()
print(some_dude.age)
```

# Error catching
Any error can be catched via a try/catch.

The code to try is inserted in the try block; The code to be executed while an error occurs is inserted in the catch block.  
The error code is also available in the catch block, via the ValueError.

```python
print("Enter your age")
try:
    age = int(input())
    if age >= 18:
        print("You are an adult")
    else:
        print("You are a minor")
except ValueError as e:
    print(f"Error: {e}")
    print("Please enter a valid integer for your age.")
```

# Interfact with files
To interact with a file, you first open it with open(), and specify how to open it (r for read, w for write, ...).  
Then you use read or write on your file object.  
You should close the file when done editing it.  

```python
filename = "example.txt"

# Write
f = open(filename, "w", encoding="utf-8") # If doesn't exist, create it.
f.write("Hello world\n")
f.write("This is a file test.\n")
f.close()

# Read
f = open(filename, "r", encoding="utf-8")
content = f.read()
f.close()

print("File content:")
print(content)
```
